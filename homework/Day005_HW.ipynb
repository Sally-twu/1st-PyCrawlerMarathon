```python
#下載檔案
#這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
webpop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = webpop.text
response
```

```python
#取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
import json
data = json.loads(response)

for d in data:
    print('title:',d['title'])
    print('createdtime:',d['createdAt'])
    print('commemtcount:',d['commentCount'])
    print('likecount:',d['likeCount'])
    print(" ")
    
print('共',len(d),'筆資料')
```

```python
#計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
webnopop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
datanopop = json.loads(webnopop.text)

commentpop = likepop = countpop = 0
for d in data:
    countpop += 1
    commentpop = commentpop+d['commentCount']
    likepop = likepop +d['likeCount']
print('熱門文章平均留言人數:',commentpop/countpop)
print('熱門文章平均按讚人數:',likepop/countpop)
print(" ")

commentnopop = likenopop = countnopop = 0
for d in datanopop:
    countnopop +=1
    commentnopop = commentnopop+d['commentCount']
    likecnopop = likenopop +d['likeCount']
print('非熱門文章平均留言人數:',commentnopop/countnopop)
print('非熱門文章平均按讚人數:',likenopop/countnopop)
```
